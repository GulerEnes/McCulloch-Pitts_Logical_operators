```python
from numpy import arange
```

```python
X = [[0, 0],
     [0, 1],
     [1, 0],
     [1, 1]]
```

```python
def F(X, Y, w0=-1, threshold=0, test_range=2, learning_rate=0.1):
    all_losses = dict()
    for w1 in arange(-test_range, test_range + learning_rate, learning_rate):
        for w2 in arange(-test_range, test_range + learning_rate, learning_rate):
            
            w1 = round(w1, 2)
            w2 = round(w2, 2)
            
            calc_Y = [0, 0, 0, 0]  # calculated y values initialization
            loss_count = 0
            
            for i in range(4):
                calc_y = w0 + w1 * X[i][0] + w2 * X[i][1]

                if calc_y > threshold:
                    calc_Y[i] = 1
                if calc_Y[i] != Y[i]:
                    loss_count += 1

            all_losses[(w1, w2)] = (loss_count, calc_Y)
            
    all_losses = sorted(list(all_losses.items()), key=lambda x: x[1][0])
    
    for k, v in all_losses:
        print(k, v)
    
    if len(all_losses) == 0:
        return "can't find"
    return all_losses[0][0]  # Returns w1 and w2 as tuple
```
